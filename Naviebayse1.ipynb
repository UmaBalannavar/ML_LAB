```python
# Load the dataset and display an image

import os
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Define the dataset path
dataset_path = 'olivetti_faces'  # Update to the correct folder path

# Example image path (adjust based on your dataset structure)
img_path = os.path.join(dataset_path, 's1', '1.pgm')  # Adjust if the folder structure is different

# Check if the file exists
if not os.path.exists(img_path):
    print(f"File not found: {img_path}")
else:
    print("File exists. Displaying the image...")

    # Load and display the image
    img = Image.open(img_path).convert('L')  # Convert to grayscale
    plt.imshow(np.array(img), cmap='gray')
    plt.axis('off')  # Hide axes
    plt.title('Example Image')
    plt.show()
```

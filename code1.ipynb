```python
import matplotlib.pyplot as plt
def fun(x):
    plt.hist(df[x],bins=10,color=['b'],alpha=0.5)
    plt.xlabel(x)
    plt.ylabel("frequency")
    plt.title("histogram of "+x)
    plt.show

def main():
    for i in df.columns:
        print(fun(i))

if __name__=="__main__" :
    main()

        
```

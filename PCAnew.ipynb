```python
import matplotlib.pyplot as plt 
import pandas as pd 
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris=load_iris()
X=iris.data
y=iris.target
```

```python
f=iris.feature_names
t=iris.target_names
plt.figure(figsize=(8,6))
for target,color,label in zip([0,1,2],['r','b','g'],t):
    plt.scatter(X[y==target,0],X[y==target,1],color=color,label=label,alpha=0.7)
plt.xlabel(f[0])  # 'sepal length (cm)'
plt.ylabel(f[1])  # 'sepal width (cm)'
plt.title('Iris Dataset: Sepal Length vs Sepal Width')
plt.legend()
plt.show()
```

```python
pca=PCA(n_components=2)
x_pca=pca.fit_transform(X)
```

```python
plt.figure(figsize=(8,6))
for target,color,label in zip([0,1,2],["red","green","blue"],iris.target_names):
    plt.scatter(x_pca[y==target,0],x_pca[y==target,1],alpha=0.7,color=color,label=label)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA")
plt.legend()
plt.show()    
```

import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Load dataset
d0 = pd.read_csv("iris.csv")

# Separate features and labels
l = d0['species']
d = d0.drop("species", axis=1)

# Standardize the data
scaler = StandardScaler()
d_scaled = scaler.fit_transform(d)

# Apply PCA (only 2 components)
pca = PCA(n_components=2)
principal_components = pca.fit_transform(d_scaled)  # Now shape will be (150, 2)

# Stack only the first two principal components with labels
pca_data = np.column_stack((principal_components, l))  # Shape will now be (150, 3)

# Create DataFrame
pca_df = pd.DataFrame(data=pca_data, columns=["1st_principal", "2nd_principal", "species"])

# Scatter plot
sn.FacetGrid(pca_df, hue="species", height=6).map(plt.scatter, "1st_principal", "2nd_principal").add_legend()
plt.title("Scatter Plot After PCA")
plt.show()
